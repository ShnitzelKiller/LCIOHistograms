```python
def analyze(name, buckets, bounds, volbounds, partbounds):
    format_md = "single%s_%d__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-25x25x25-blur_4-buffer_25-threshold_1-volthreshold_10-%s.txt"
    gamma_md1 = loadtxt(format_md % (name, 1, name, "meshdensity"))
    gamma_md2 = loadtxt(format_md % (name, 2, name, "meshdensity"))
    gamma_md5 = loadtxt(format_md % (name, 5, name, "meshdensity"))
    gamma_md10 = loadtxt(format_md % (name, 10, name, "meshdensity"))
    gamma_md20 = loadtxt(format_md % (name, 20, name, "meshdensity"))
    gamma_md50 = loadtxt(format_md % (name, 50, name, "meshdensity"))
    gamma_md100 = loadtxt(format_md % (name, 100, name, "meshdensity"))
    
    gamma_hd1 = loadtxt(format_md % (name, 1, name, "hulldensity"))
    gamma_hd2 = loadtxt(format_md % (name, 2, name, "hulldensity"))
    gamma_hd5 = loadtxt(format_md % (name, 5, name, "hulldensity"))
    gamma_hd10 = loadtxt(format_md % (name, 10, name, "hulldensity"))
    gamma_hd20 = loadtxt(format_md % (name, 20, name, "hulldensity"))
    gamma_hd50 = loadtxt(format_md % (name, 50, name, "hulldensity"))
    gamma_hd100 = loadtxt(format_md % (name, 100, name, "hulldensity"))
    
    fig = figure()
    ax1 = fig.add_subplot(211)
    ax1.hist(log(gamma_md1[gamma_md1 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md2[gamma_md2 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md5[gamma_md5 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md10[gamma_md10 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md20[gamma_md20 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md50[gamma_md50 > 0]), buckets, bounds)
    ax1.hist(log(gamma_md100[gamma_md100 > 0]), buckets, bounds)
    title("Mesh Energy Density")
    #legend(["1", "2", "5", "10", "20", "50", "100"], loc="best")
    ax2 = fig.add_subplot(212)
    ax2.hist(log(gamma_hd1[gamma_hd1 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd2[gamma_hd2 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd5[gamma_hd5 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd10[gamma_hd10 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd20[gamma_hd20 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd50[gamma_hd50 > 0]), buckets, bounds)
    ax2.hist(log(gamma_hd100[gamma_hd100 > 0]), buckets, bounds)
    title("Hull Energy Density")
    #legend(["1", "2", "5", "10", "20", "50", "100"], loc="best")
    
    gamma_vr1 = loadtxt(format_md % (name, 1, name, "volumeratio"))
    gamma_vr2 = loadtxt(format_md % (name, 2, name, "volumeratio"))
    gamma_vr5 = loadtxt(format_md % (name, 5, name, "volumeratio"))
    gamma_vr10 = loadtxt(format_md % (name, 10, name, "volumeratio"))
    gamma_vr20 = loadtxt(format_md % (name, 20, name, "volumeratio"))
    gamma_vr50 = loadtxt(format_md % (name, 50, name, "volumeratio"))
    gamma_vr100 = loadtxt(format_md % (name, 100, name, "volumeratio"))
    
    fig2 = figure()
    ax3 = fig2.add_subplot(111)
    ax3.hist(gamma_vr1, buckets, volbounds)
    ax3.hist(gamma_vr2, buckets, volbounds)
    ax3.hist(gamma_vr5, buckets, volbounds)
    ax3.hist(gamma_vr10, buckets, volbounds)
    ax3.hist(gamma_vr20, buckets, volbounds)
    ax3.hist(gamma_vr50, buckets, volbounds)
    ax3.hist(gamma_vr100, buckets, volbounds);
    title("Volume Ratio")
    legend(["1", "2", "5", "10", "20", "50", "100"], loc="best")
    
    mvs1 = loadtxt(format_md % (name, 1, name, "mesheccentricity"))
    mvs2 = loadtxt(format_md % (name, 2, name, "mesheccentricity"))
    mvs5 = loadtxt(format_md % (name, 5, name, "mesheccentricity"))
    mvs10 = loadtxt(format_md % (name, 10, name, "mesheccentricity"))
    mvs20 = loadtxt(format_md % (name, 20, name, "mesheccentricity"))
    mvs50 = loadtxt(format_md % (name, 50, name, "mesheccentricity"))
    mvs100 = loadtxt(format_md % (name, 100, name, "mesheccentricity"))
    
    hvs1 = loadtxt(format_md % (name, 1, name, "hulleccentricity"))
    hvs2 = loadtxt(format_md % (name, 2, name, "hulleccentricity"))
    hvs5 = loadtxt(format_md % (name, 5, name, "hulleccentricity"))
    hvs10 = loadtxt(format_md % (name, 10, name, "hulleccentricity"))
    hvs20 = loadtxt(format_md % (name, 20, name, "hulleccentricity"))
    hvs50 = loadtxt(format_md % (name, 50, name, "hulleccentricity"))
    hvs100 = loadtxt(format_md % (name, 100, name, "hulleccentricity"))
    
    fig3 = figure()
    ax5a = fig3.add_subplot(211)
    ax5a.hist(mvs1, buckets, volbounds)
    ax5a.hist(mvs2, buckets, volbounds)
    ax5a.hist(mvs5, buckets, volbounds)
    ax5a.hist(mvs10, buckets, volbounds)
    ax5a.hist(mvs20, buckets, volbounds)
    ax5a.hist(mvs50, buckets, volbounds)
    ax5a.hist(mvs100, buckets, volbounds)
    title("Mesh Volume/Surface Area")
    ax5b = fig3.add_subplot(212)
    ax5b.hist(hvs1, buckets, volbounds)
    ax5b.hist(hvs2, buckets, volbounds)
    ax5b.hist(hvs5, buckets, volbounds)
    ax5b.hist(hvs10, buckets, volbounds)
    ax5b.hist(hvs20, buckets, volbounds)
    ax5b.hist(hvs50, buckets, volbounds)
    ax5b.hist(hvs100, buckets, volbounds)
    title("Hull Volume/Surface Area")
    
    mps1 = loadtxt(format_md % (name, 1, name, "meshparts"))
    mps2 = loadtxt(format_md % (name, 2, name, "meshparts"))
    mps5 = loadtxt(format_md % (name, 5, name, "meshparts"))
    mps10 = loadtxt(format_md % (name, 10, name, "meshparts"))
    mps20 = loadtxt(format_md % (name, 20, name, "meshparts"))
    mps50 = loadtxt(format_md % (name, 50, name, "meshparts"))
    mps100 = loadtxt(format_md % (name, 100, name, "meshparts"))
    
    fig5 = figure()
    ax6 = fig5.add_subplot(111)
    ax6.hist(mps1, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps2, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps5, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps10, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps20, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps50, partbounds[1]-partbounds[0], partbounds)
    ax6.hist(mps100, partbounds[1]-partbounds[0], partbounds)
    title("Number of Mesh Parts with $>10\%$ Max Volume")
    
    energies1 = loadtxt("single%s_1__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies2 = loadtxt("single%s_2__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies5 = loadtxt("single%s_5__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies10 = loadtxt("single%s_10__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies20 = loadtxt("single%s_20__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies50 = loadtxt("single%s_50__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    energies100 = loadtxt("single%s_100__scint1x1_%s_theta90_phi90_REC-PandoraPFOCollection-events_1_to_1000-MCEnergy.txt" % (name, name))
    
    energy1 = mean(energies1)
    std1 = std(energies1)
    energy2 = mean(energies2)
    std2 = std(energies2)
    energy5 = mean(energies5)
    std5 = std(energies5)
    energy10 = mean(energies10)
    std10 = std(energies10)
    energy20 = mean(energies20)
    std20 = std(energies20)
    energy50 = mean(energies50)
    std50 = std(energies50)
    energy100 = mean(energies100)
    std100 = std(energies100)
    
    fig3 = figure()
    ax4 = fig3.add_subplot(111)
    ax4.fill_between([1, 2, 5, 10, 20, 50, 100], [energy1-std1, energy2-std2, energy5-std5, energy10-std10, energy20-std20, energy50-std50, energy100-std100], [energy1+std1, energy2+std2, energy5+std5, energy10+std10, energy20+std20, energy50+std50, energy100+std100], color=(1, 0.5, 0.5))
    xlabel("Actual Energy (GeV)")
    ylabel("Reconstructed Energy (GeV)")
    title("Reconstructed Energy")
    
```
